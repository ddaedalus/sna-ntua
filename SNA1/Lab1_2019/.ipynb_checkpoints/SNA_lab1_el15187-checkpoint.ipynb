### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

##### 2)  Να αναπαρασταθεί (plot) η κατανομή βαθμών κόμβων του δικτύου (δηλαδή, για κάθε βαθμό πόσοι κόμβοι έχουν αυτό το βαθμό) καθώς και η συγκεντρωτική κατανομή βαθμού κόμβου (δηλαδή για κάθε βαθμό πόσοι κόμβοι έχουν το πολύ αυτό το βαθμο)

```python
#για την κατανομή βαθμού κόμβου, χρησιμοποιήστε την hist του matplotlib

# graph_degree: dict{pos} = degree
def list_node_degrees(graph_degree):
    return list(graph_degree.values())

def find_discriminant_degrees(graph_degree):
    degrees = list_node_degrees(graph_degree)
    max_degree = np.max(degrees)
    # Frequency Table of degrees
    freq_degree = []
    for i in range(max_degree + 1):
        freq_degree.append(0)
        
    for degree in degrees:
        freq_degree[degree] += 1
            
    x_hist = []
    for i in range(len(freq_degree)):
        if (freq_degree[i] > 0):
            for j in range(freq_degree[i]):
                x_hist.append(i)
    return x_hist
    
plt.figure(1)
plt.title("κατανομη βαθμου κομβων του REG")
plt.hist(find_discriminant_degrees(reg_degree))
plt.figure(2)
plt.title("κατανομη βαθμου κομβων του RGER")
plt.hist(find_discriminant_degrees(rger_degree))
plt.figure(3)
plt.title("κατανομη βαθμου κομβων του RGG")
plt.hist(find_discriminant_degrees(rgg_degree))
plt.figure(4)
plt.title("κατανομη βαθμου κομβων του SF")
plt.hist(find_discriminant_degrees(sf_degree))
plt.figure(5)
plt.title("κατανομη βαθμου κομβων του SW")
plt.hist(find_discriminant_degrees(sw_degree))

#για τη συγκεντρωτική κατανομή του διανύσματος values (σε αυτή την περίπτωση τa degree values),
#χρησιμοποιήστε την plot(X,Y) του matplotlib
Xreg = np.sort(list(reg_degree.values())) 
Xrger = np.sort(list(rger_degree.values())) 
Xrgg = np.sort(list(rgg_degree.values())) 
Xsf = np.sort(list(sf_degree.values())) 
Xsw = np.sort(list(sw_degree.values())) 
Y = np.array(range(n))/float(n)
plt.figure(6)
plt.title("συγκεντρωτικη κατανομη βαθμου κομβου του REG")
plt.plot(Xreg, Y)
plt.figure(7)
plt.title("συγκεντρωτικη κατανομη βαθμου κομβου του RGER")
plt.plot(Xrger, Y)
plt.figure(8)
plt.title("συγκεντρωτικη κατανομη βαθμου κομβου του RGG")
plt.plot(Xrgg, Y)
plt.figure(9)
plt.title("συγκεντρωτικη κατανομη βαθμου κομβου του SF")
plt.plot(Xsf, Y)
plt.figure(10)
plt.title("συγκεντρωτικη κατανομη βαθμου κομβου του SW")
plt.plot(Xsw, Y)
```

### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx
a = 1
b = 10

def array_line(n):
    return [np.random.rand()*(b-a) + a for i in range(n)]

def adjacent_array(n):
    # Adjacent Weighted Array
    weight_array = np.array([array_line(n) for i in range(n)])
    # Symmetric Array
    symmetric = (np.maximum(weight_array, weight_array.transpose()))
    # (i,i) = 0
    for i in range(n):
        symmetric[i][i] = 0
    
    return symmetric
    
# Adding weighted edge
def addWeightedEdge(G, n):
    adj = adjacent_array(n)
    for e in G.edges():
        n0 = e[0]
        n1 = e[1]
        G[n0][n1]['weight'] = adj[n0][n1]
    return G
        
reg_weighted = addWeightedEdge(REG, n)

rger_weighted = addWeightedEdge(RGER, n)

rgg_weighted = addWeightedEdge(RGG, n)

sf_weighted = addWeightedEdge(SF, n)

sw_weighted = addWeightedEdge(SW, n)
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου
**Node Strength**: The sum of weights attached to ties belonging to a node 

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(G):
    weights_list = []
    for i in range(len(G.adj)):
        a = []
        for j in range(len(G.adj[i].items())):
            d = list(G.adj[i].values())[j]['weight']   
            a.append(d)
        weights_list.append(a)    
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength = strength(reg_weighted)
RGER_strength = strength(rger_weighted)
RGG_strength = strength(rgg_weighted)
SF_strength = strength(sf_weighted)
SW_strength = strength(sw_weighted)
print(REG_strength)
print()

print(RGER_strength)
print()

print(RGG_strength)
print()

print(SF_strength)
print()

print(SW_strength)
print()
```

##### ii) Να αναπαρασταθεί η συγκεντρωτική κατανομή δύναμης (cumulative strength distribution) των κόμβων του δικτύου (σε κάθε τιμή δύναμης αντιστοιχίζεται το πλήθος των κόμβων με δύναμη μικρότερη ή ίση από αυτή)

```python
# Οπως Β (ii)
Xreg = np.sort(list(REG_strength)) 
Xrger = np.sort(list(RGER_strength))
Xrgg = np.sort(list(RGG_strength)) 
Xsf = np.sort(list(SF_strength)) 
Xsw = np.sort(list(SW_strength))
Y = np.array(range(n))/float(n)
plt.figure(6)
plt.title("συγκεντρωτικη κατανομη δυναμης REG")
plt.plot(Xreg, Y)

plt.figure(7)
plt.title("συγκεντρωτικη κατανομη δυναμης RGER")
plt.plot(Xrger, Y)

plt.figure(8)
plt.title("συγκεντρωτικη κατανομη δυναμης RGG")
plt.plot(Xrgg, Y)

plt.figure(9)
plt.title("συγκεντρωτικη κατανομη δυναμης SF")
plt.plot(Xsf, Y)

plt.figure(10)
plt.title("συγκεντρωτικη κατανομη δυναμης SW")
plt.plot(Xsw, Y)
```

### Δ) Υπολογισμός μέσου μήκους μονοπατιού

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του στο δίκτυο. Να παρουσιαστούν συγκεντρωτικά τα αποτελέσματα. Πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως εξηγήθηκαν στη θεωρία.

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx 
def compute_path_lengths(G):
    return nx.all_pairs_shortest_path_length(G)

REG_path_lengths = list(compute_path_lengths(REG))
RGER_path_lengths = list(compute_path_lengths(RGER))
RGG_path_lengths = list(compute_path_lengths(RGG))
SF_path_lengths = list(compute_path_lengths(SF))
SW_path_lengths = list(compute_path_lengths(SW))

def getLengths(G_path_lengths):
    lengths = []
    for item in G_path_lengths:
        item_dict = item[1]
        for value in item_dict.values():
            lengths.append(value)
            
    return lengths

REG_lengths = getLengths(REG_path_lengths)
RGER_lengths = getLengths(RGER_path_lengths)
RGG_lengths = getLengths(RGG_path_lengths)
SF_lengths = getLengths(SF_path_lengths)
SW_lengths = getLengths(SW_path_lengths)

print("Mean path length REG: " + str(np.mean(REG_lengths)))
print("Variance of path length REG: " + str(np.var(REG_lengths)))
print()

print("Mean path length RGER: " + str(np.mean(RGER_lengths)))
print("Variance of path length RGER: " + str(np.var(RGER_lengths)))
print()

print("Mean path length RGG: " + str(np.mean(RGG_lengths)))
print("Variance of path length RGG: " + str(np.var(RGG_lengths)))
print()

print("Mean path length SF: " + str(np.mean(SF_lengths)))
print("Variance of path length SF: " + str(np.var(SF_lengths)))
print()

print("Mean path length SW: " + str(np.mean(SW_lengths)))
print("Variance of path length SW: " + str(np.var(SW_lengths)))
```

### Ε) Υπολογισμός συντελεστή ομαδοποίησης (ΣΟ)

The **local clustering coefficient** $C_i$ for a vertex $u_i$ is given by the proportion of links between the vertices within its neighbourhood divided by the number of links that could possibly exist between them.

##### 1) Αναλυτικός υπολογισμός του ΣΟ

Για κάθε μια από τις παρακάτω τοπολογίες να υπολογιστεί αναλυτικά (να φαίνονται οι πράξεις) το τοπικό ΣΟ του κάθε κόμβου και να δειχθεί η κατανομή του τοπικού ΣΟ για όλους τους κόμβους.

Στη συνέχεια να υπολογιστεί και να συγκριθεί ο μέσος ΣΟ κάθε τοπολογίας.

```python
## Προσοχή! Ο παρακάτω κώδικας παράγει τις τοπολογίες στις οποίες θα κάνετε αναλυτικούς υπολογισμούς.
## ΜΗΝ ΤΟΝ ΑΛΛΑΞΕΤΕ

plt.figure(1, figsize=(2,2))
plt.title('Topology 1')
G1 = nx.Graph()
pos = {0: (1, 1), 1: (0, 1), 2: (1, 0), 3: (2, 1), 4: (1, 2)}
G1.add_nodes_from(pos.keys())
G1.add_edges_from([(0,1),(0,2),(0,3),(0,4)])
labels = {}
labels[0] = r'$0$'
labels[1] = r'$1$'
labels[2] = r'$2$'
labels[3] = r'$3$'
labels[4] = r'$4$'
nx.draw_networkx_labels(G1, pos, labels)
nx.draw(G1,pos)
# Dict of clustering coefficient for each node
co1 = nx.clustering(G1)
print("Topology 1, clustering coefficient for each node:")
print(co1)
print("Average clustering coefficient for Topology1: " + str( np.mean(list(co1.values()) )))
print()

plt.figure(2,figsize=(2,2))
plt.title('Topology 2')
G2 = nx.Graph()
pos = {0: (1, 1), 1: (0, 0), 2: (2, 0), 3: (2, 2), 4: (0, 2)}
G2.add_nodes_from(pos.keys())
G2.add_edges_from([(0,1),(0,2),(0,3),(0,4),(1,2),(3,4)])
labels = {}
labels[0] = r'$0$'
labels[1] = r'$1$'
labels[2] = r'$2$'
labels[3] = r'$3$'
labels[4] = r'$4$'
nx.draw_networkx_labels(G2, pos, labels)
nx.draw(G2, pos)
# Dict of clustering coefficient for each node
co2 = nx.clustering(G2)
print("Topology 2, clustering coefficient for each node:")
print(co2)
print("Average clustering coefficient for Topology2: " + str( np.mean(list(co2.values()) )))
print()

plt.figure(3,figsize=(2,2))
plt.title('Topology 3')
G3 = nx.Graph()
pos = {0: (1, 1), 1: (0, 1), 2: (1, 0), 3: (2, 1), 4: (1, 2)}
G3.add_nodes_from(pos.keys())
G3.add_edges_from([(0,1),(0,2),(0,3),(0,4),(1,2),(2,3),(3,4),(4,1)])
labels = {}
labels[0] = r'$0$'
labels[1] = r'$1$'
labels[2] = r'$2$'
labels[3] = r'$3$'
labels[4] = r'$4$'
nx.draw_networkx_labels(G3, pos, labels)
nx.draw(G3,pos)
# Dict of clustering coefficient for each node
co3 = nx.clustering(G3)
print("Topology 3, clustering coefficient for each node:")
print(co3)
print("Average clustering coefficient for Topology3: " + str( np.mean(list(co3.values()) )))
print()
```

### Ζ) Υπολογισμός κεντρικότητας κόμβων

**Degree centrality** of a node refers to the number of edges attached to the node. In order to know the standardized score, you need to divide each score by $n-1$ (n: the number of nodes).  
**Closeness centrality** (or closeness) of a node is a measure of centrality in a network, calculated as the reciprocal of the sum of the length of the shortest paths between the node and all other nodes in the graph. Thus, the more central a node is, the closer it is to all other nodes.  
\begin{align*}
C(x) = \frac{1}{\sum_y d(y,x)}
\end{align*}  
**Betweenness centrality** is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. Betweenness centrality is a measure of centrality in a graph based on shortest paths. For every pair of vertices in a connected graph, there exists at least one shortest path between the vertices such that either the number of edges that the path passes through (for unweighted graphs) or the sum of the weights of the edges (for weighted graphs) is minimized. The betweenness centrality for each vertex is the number of these shortest paths that pass through the vertex.  
###### - σ(s,t)
Number of shortest paths between nodes s and t  
###### - σ(s,t|v)
Number of shortest paths between nodes s and t that pass through v. 
  
  
\begin{align*}  
C_B(v) = \frac{\sum_s \sum_t σ(s,t|v)}{\sum_s \sum_t σ(s,t)}, 
\end{align*}  
**Eigenvector centrality** (also called eigencentrality) is a measure of the influence of a node in a network. It assigns relative scores to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes.

### Ι) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

Μεταβάλλοντας την παράμετρο $g_p$ της συνάρτησης που κατασκευάζει τοπολογίες SW, από $0$ έως $1$ με βήμα $0.1$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγώμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
n = 170
d = 4

def compute_path_lengths(G):
    return nx.all_pairs_shortest_path_length(G)


def getLengths(G_path_lengths):
    lengths = []
    for item in G_path_lengths:
        item_dict = item[1]
        for value in item_dict.values():
            lengths.append(value)
            
    return lengths


def calculate_clust_coef(G):
    return nx.clustering(G)


for p in list(np.linspace(0, 1, 11)):
    p = int(p*10) / 10
    if (p == 1 or p == 0):
        p = int(p)
    sw = nx.watts_strogatz_graph(n, d, p)
    sw_path_lengths = list(compute_path_lengths(sw))
    sw_lengths = getLengths(SW_path_lengths)
    sw_clust_coef = calculate_clust_coef(sw)
    print("Mean path length, p = " + str(p) + ": \t\t" +  str(np.mean(sw_lengths)))
    print("Mean clustering coefficient: \t\t" + str(np.mean(list(sw_clust_coef.values()))))
    print()
```
