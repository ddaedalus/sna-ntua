<b>Σχολιασμός αποτελεσμάτων</b>  
Ξεκινώντας απο το REG, εφόσον είναι πεπερασμένο, δεν έχει ουδεμία τυχαιότητα, αφού ο βαθμός όλων των κόμβων πρέπει οπωσδήποτε να είναι ο ίδιος, δηλαδή ίσον με 4. Με τον ίδιο τρόπο, περιμένουμε η μέση τιμή των βαθμών των κόμβων να είναι και αυτή ίση με 4 και η διασπορά προφανώς ίση με το 0. Έτσι, η κατανομή του βαθμού των κόμβων αποτελεί την Dirac function.  

Για το Erdos-Renyi (RGER) μοντέλο, o γράφος διαλέγεται uniformly τυχαία από το σύνολο όλων των γράφων που έχουν Ν κόμβους και Μ ακμές. Έτσι, η πιθανότητα ένας κόμβος να έχει Κ βαθμό ακολουθεί την Poisson distribution. H μέση τιμή των βαθμών κόμβων περιμένουμε να είναι περίπου ίση με το $p\cdot(Ν-1)$, όπου p η πιθανότητα ένας γράφος να διαλεχτεί από το πλήθος των γράφων που ορίζει το μοντέλο. Αύτο έχει ως αποτέλεσμα, να περιμένουμε την διασπορά να έχει την ίδια τιμή με την μέση τιμή, κάτι το οποίο εν μέρει ισχύει και στο πείραμά μας, αφού οι δύο τιμές έχουν πολύ μικρή απόκλιση.  

Στο RGG μοντέλο, έχουμε εναν μη κατευθυνόμενο γεωμετρικό γράφο, με κόμβους τυχαία παρμένους από μία ομοιόμορφη κατανομή του χώρου $[0,1)^2$. Δύο κόμβοι είναι συνδεδεμένοι αν και μόνο αν η ευκλίδεια απόστασή τους είναι μικρότερη από μία προκαθορισμένη παράμετρο $r$ $ε$ $(0,1)$. Έτσι, περιμένουμε η κατανομή του βαθμού κόμβου να ακολουθεί την uniform distribution, όπως και παρατηρούμε στο πείραμά μας εν μέρει να προσομοιάζει, με σχετικά μεγάλη απόκλιση της διακύμανσης από την μέση τιμή. Περιμένουμε, σύμφωνα με την θεωρία, ο μέσος βαθμός κόμβου να είναι $nπ r^2$, θεωρητικό αποτέλεσμα που στο πείραμά μας είναι κοντά με το πραγματικό.   

Στο Barabasi-Albert (SF) μοντέλο, το δίκτυο αρχικοποιείται απο έναν αρχικό συνδεδεμένο γράφο που εχει $Μ_0$ κόμβους. Στην συνέχεία, νέοι κόμβοι προσθέτονται, ένας κάθε φορά. Κάθε νέος κόμβος συνδέεται με $ Μ \leq Μ_0$ υπάρχοντες κόμβους με πιθανότητα ανάλογη του αριθμού των συνδέσεων που οι υπάρχοντες κόμβοι ήδη έχουν.  
Πιο αναλυτικά:  
$$p_i = \frac {k_i} {\sum_{j} k_j}$$  
Έτσι, οι νέοι κόμβοι έχουν την προτίμηση να συνδέονται με τους πιο heavily υπάρχοντες linked κόμβους. Όπως, παρατηρούμε και στο πείραμά μας, η κατανομή των κόμβων ακολούθει power-law distribution. Αυτό εξηγεί και την μεγάλη διασπορά που έχουμε, που αν και πεπερασμένη, αποκλίνει πολύ σε σχέση με την μέση τιμή του βαθμού κόμβου. Με αυτόν τον τρόπο, καταφέρνουμε να συνυπάρχουμε λίγους κόμβους μεγάλου βαθμού με πολλούς κόμβους μικρού βαθμού.


Τέλος, στο Watts-Strogatz (SW) μοντέλο, δοθέντους του Ν, αριθμού κόμβων, του μέσου βαθμού κόμβου Κ και της ειδικής παραμέτρου β, κατασκευάζεται ένας μη κατευθυνόμενος γράφος με Ν κόμβους και $\frac{NK}{2}$ ακμές. Γνωρίζουμε ότι η μέση τιμή βαθμού κόμβου πρέπει να παίρνει την τιμή του Κ που δώσαμε, κάτι που ικανοποιείται επιτυχώς και στο πείραμά μας. Επιπλέον, η κατανομή του βαθμού κόμβου φαίνεται να ακολουθεί την heavy-tail distribution, που σημαίνει στο πείραμά μας ότι οι περισσότεροι κόμβοι τείνουν να έχουν βαθμό μεγάλης τιμής, δηλαδή κοντά στην μέση τιμή Κ, εν προκειμένου ίση με 4, και οι εναπομείναντες να συμπληρώνονται μεταξύ τους με τιμές κάτω ή πάνω απο το Κ.


### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx
a = 1
b = 10

def array_line(n):
    return [np.random.rand()*(b-a) + a for i in range(n)]

def adjacent_array(n):
    # Adjacent Weighted Array
    weight_array = np.array([array_line(n) for i in range(n)])
    # Symmetric Array
    symmetric = (np.maximum(weight_array, weight_array.transpose()))
    # (i,i) = 0
    for i in range(n):
        symmetric[i][i] = 0
    
    return symmetric
    
# Adding weighted edge
def addWeightedEdge(G, n):
    adj = adjacent_array(n)
    for e in G.edges():
        n0 = e[0]
        n1 = e[1]
        G[n0][n1]['weight'] = adj[n0][n1]
    return G
        
reg_weighted = addWeightedEdge(REG, n)

rger_weighted = addWeightedEdge(RGER, n)

rgg_weighted = addWeightedEdge(RGG, n)

sf_weighted = addWeightedEdge(SF, n)

sw_weighted = addWeightedEdge(SW, n)
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου
**Node Strength**: The sum of weights attached to ties belonging to a node 

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(G):
    weights_list = []
    for i in range(len(G.adj)):
        a = []
        for j in range(len(G.adj[i].items())):
            d = list(G.adj[i].values())[j]['weight']   
            a.append(d)
        weights_list.append(a)    
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength = strength(reg_weighted)
RGER_strength = strength(rger_weighted)
RGG_strength = strength(rgg_weighted)
SF_strength = strength(sf_weighted)
SW_strength = strength(sw_weighted)
print("REG Strength")
print(REG_strength)
print()

print("RGER Strength")
print(RGER_strength)
print()

print("RGG Strength")
print(RGG_strength)
print()

print("SF Strength")
print(SF_strength)
print()

print("SW Strength")
print(SW_strength)
print()
```

Για το RGG μοντέλο, αναλόγως την τιμή που έχουμε δώσει στο r, περιμένουμε ο κάθε κόμβος να συνδέεται με λίγους ή πολλούς άλλους κόμβους. Στο πείραμά μας, προφανώς, η τιμή του r επιτρέπει μεγάλο μέσο βαθμό κόμβου και άρα μεγάλη μέση δύναμη. 

Για το SF μοντέλο, επείδη υπάρχουν λίγοι κόμβοι με μεγάλο βαθμό κόμβου και πολλοί με μικρό, αυτό συνεπάγεται ότι θα υπάρχουν λίγοι κόμβοι με πολύ μεγάλη δύναμη και πολλοί με μικρή. Έτσι, περιμένουμε τεράστια διασπορά στην δύναμη του γράφου και σχετικά μεγάμη μέση δύναμη.

Τέλος για το SW μοντέλο, περιμένουμε παρόμοια μέση τιμή δύναμης με το REG μοντέλο, αφού έχουν το ίδιο mean degree, $Κ=4$, ενώ περιμένουμε σημαντική διασπορά δύναμης διότι το SW degree ακολουθεί την heavy-tailed distribution και άρα αρκετοί κόμβοι θα έχουν δύναμη που θα αποκλίνει από την μέση τιμή.

### Δ) Υπολογισμός μέσου μήκους μονοπατιού

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του στο δίκτυο. Να παρουσιαστούν συγκεντρωτικά τα αποτελέσματα. Πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως εξηγήθηκαν στη θεωρία.

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx 
def compute_path_lengths(G):
    return nx.all_pairs_shortest_path_length(G)

REG_path_lengths = list(compute_path_lengths(REG))
RGER_path_lengths = list(compute_path_lengths(RGER))
RGG_path_lengths = list(compute_path_lengths(RGG))
SF_path_lengths = list(compute_path_lengths(SF))
SW_path_lengths = list(compute_path_lengths(SW))

def getLengths(G_path_lengths):
    lengths = []
    for item in G_path_lengths:
        item_dict = item[1]
        for value in item_dict.values():
            lengths.append(value)
            
    return lengths

REG_lengths = getLengths(REG_path_lengths)
RGER_lengths = getLengths(RGER_path_lengths)
RGG_lengths = getLengths(RGG_path_lengths)
SF_lengths = getLengths(SF_path_lengths)
SW_lengths = getLengths(SW_path_lengths)

print("Mean path length REG: " + str(np.mean(REG_lengths)))
print("Variance of path length REG: " + str(np.var(REG_lengths)))
print()

print("Mean path length RGER: " + str(np.mean(RGER_lengths)))
print("Variance of path length RGER: " + str(np.var(RGER_lengths)))
print()

print("Mean path length RGG: " + str(np.mean(RGG_lengths)))
print("Variance of path length RGG: " + str(np.var(RGG_lengths)))
print()

print("Mean path length SF: " + str(np.mean(SF_lengths)))
print("Variance of path length SF: " + str(np.var(SF_lengths)))
print()

print("Mean path length SW: " + str(np.mean(SW_lengths)))
print("Variance of path length SW: " + str(np.var(SW_lengths)))
```

Για το RGG model, περιμένουμε μεγάλο μέσο μήκος μονοπατιού, κάτι το οποίο, ωστόσο, εξαρτάται από την Ευκλίδεια απόσταση μεταξύ των κόμβων στον $R^d$, εδώ $R^2$, και την τιμή της παραμέτρου r που έχουμε ορίσει από προηγουμένως. Για το SF model, περιμένουμε πολύ μικρό μήκος μέσο μήκος μονοπατιού της τάξης του $\frac {lnN} {ln(lnN)}$. Παίρνουμε, ωστόσο, τιμή μικρότερη της αναμενόμενης κατα μία μονάδα περίπου. Επίσης, λαμβάνουμαι πάρα πολύ μικρή διασπορά για το μέσο μήκος μονοπατιού, γεγονός που εξηγεί πως οι περισσότεροι κόμβοι είτε είναι κόμβοι μεγάλου degree είτε συνδέονται με ένα hop με κόμβο μεγάλου degree.

### Ζ) Υπολογισμός κεντρικότητας κόμβων

**ΟΡΙΣΜΟΙ**

**Degree centrality** of a node refers to the number of edges attached to the node. In order to know the standardized score, you need to divide each score by $n-1$ (n: the number of nodes).  
**Closeness centrality** (or closeness) of a node is a measure of centrality in a network, calculated as the reciprocal of the sum of the length of the shortest paths between the node and all other nodes in the graph. Thus, the more central a node is, the closer it is to all other nodes.  
\begin{align*}
C(x) = \frac{1}{\sum_y d(y,x)}
\end{align*}  
**Betweenness centrality** is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. Betweenness centrality is a measure of centrality in a graph based on shortest paths. For every pair of vertices in a connected graph, there exists at least one shortest path between the vertices such that either the number of edges that the path passes through (for unweighted graphs) or the sum of the weights of the edges (for weighted graphs) is minimized. The betweenness centrality for each vertex is the number of these shortest paths that pass through the vertex.  
###### - σ(s,t)
Number of shortest paths between nodes s and t  
###### - σ(s,t|v)
Number of shortest paths between nodes s and t that pass through v. 
  
  
\begin{align*}  
C_B(v) = \frac{\sum_s \sum_t σ(s,t|v)}{\sum_s \sum_t σ(s,t)}, 
\end{align*}  
**Eigenvector centrality** (also called eigencentrality) is a measure of the influence of a node in a network. It assigns relative scores to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes.

Για κάθε έναν τύπο δικτύου, να παράξετε 100 διαφορετικές τοπολογίες και να παρουσιάσετε το διάγραμμα ποσοστού συνεκτικότητας σε σχέση με τις καθορισμένες παραμέτρους κάθε δικτύου.

```python
#για τον έλεγχο της συνεκτικότητας, χρησιμοποίηστε την is_connected του networkx
#για τη γραφική αναπαράσταση των αποτελεσμάτων, μπορείτε να χρησιμοποίησετε την bar του matplotlib
#έλεγχος συνεκτικότητας για REG

# Επιστρεφω το ποσοστο των συνδεδεμενων γραφηματων
def REG_connect_per(d, n, fn):
    connected = 0
    for i in range(100): # 100 different topologies
        tmp_REG = fn(n,d)
        connected = connected + nx.is_connected(tmp_REG)
    return connected / 100
        
    
#plot example:
def plot_connect_REG(n, idx, fn):
    plt.figure(idx)
    y = [REG_connect_per(d, n, fn) for d in range(2,12,2)]
    plt. bar(list(range(2,12,2)),y, width=0.5, color='r')
    plt.title("REG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree")
    plt.show()
    
reg_fn = lambda n,d:nx.watts_strogatz_graph(n, d, 0)
plot_connect_REG(100, 1, reg_fn)
plot_connect_REG(200, 2, reg_fn)
```

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστείτε τις adjacency_matrix (networkx), to.dense, ones (numpy)
#linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.
```

```python
d = 4
# Lattice -> p = 0
REG =  nx.watts_strogatz_graph(n, d, 0, seed=None)

M = 750
# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)

R = 0.25
L = 1
n = 170
# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, dim=2) 

d = 4
# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n, d)

n = 170
d = 4
p = 0.3
# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)

def ego_central(topology, name):
    ego_centrality_list = []
    for node in REG.nodes():
        node_ego = nx.ego_graph(topology, node)
        # adjacency matrix
        A = (nx.adj_matrix(node_ego)).todense()
        # A^2 * (1-A)
        hadamard = np.multiply(np.matmul(A,A), (1-A))
        triu = np.triu(hadamard, 1)
        triu = triu[triu > 0]
        ego_betweenness = np.sum([1/i for i in triu])
        ego_centrality_list.append(ego_betweenness / 2)

    print("Ego-centralities for each node in " + name + ": \n" + str(ego_centrality_list) + "\n")
    return ego_centrality_list
    
ego_reg = ego_central(REG, "REG")
ego_rger = ego_central(RGER, "RGER")
ego_rgg = ego_central(RGG, "RGG")
ego_sf = ego_central(SF, "SF")
ego_sw = ego_central(SW, "SW")
```
